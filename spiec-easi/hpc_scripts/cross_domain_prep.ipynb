```python
import pandas as pd
```

```python
ids = pd.read_excel("Methylation_IDs.xlsx")
betas = pd.read_csv("clock_betas.csv")
```

```python
# Remove rows where 'Sentrix_ID' contains NaN values
ids = ids.dropna(subset=['Sentrix_ID'])

ids
```

```python
betas
```

```python
df = pd.read_csv("../ML_Data/FINAL_GENUS_TAXA_CLR.csv")
```

```python
# Function to safely format 'Sentrix_ID' without altering its value
def format_sentrix_id(value):
    if pd.isna(value):
        return None  # Return None if value is NaN
    return str(int(value)) if value == int(value) else str(value)

# Apply the formatting function to 'Sentrix_ID'
ids['Sentrix_ID'] = ids['Sentrix_ID'].apply(format_sentrix_id)

# Create the `Sentrix_Identifier` column
ids['Sentrix_Identifier'] = (ids['Sentrix_ID'] + "_" + ids['Sentrix_Position']).astype(str)

# Verify the changes
print(ids[['Sentrix_ID', 'Sentrix_Identifier']].head())
```

```python
import pandas as pd

# Assuming `ids`, `betas`, and `df` are your DataFrames
# Replace column names in `betas` based on the matching conditions

# Iterate over the columns in `betas` and replace names based on the conditions
new_columns = []
for col in betas.columns:
    # Check if the column name matches any Sentrix_Identifier in `ids`
    match = ids[ids['Sentrix_Identifier'] == col]
    if not match.empty:
        # Check which ID to use (True ID, Alternative ID, or default)
        true_id = match['True ID'].values[0]
        alternative_id = match['Alternative ID'].values[0]
        
        if true_id in df['Sample_ID'].values:
            new_columns.append(true_id)
        elif alternative_id in df['Sample_ID'].values:
            new_columns.append(alternative_id)
        else:
            new_columns.append(col)  # Keep the original column name
    else:
        new_columns.append(col)  # Keep the original column name

# Assign the new column names to the betas DataFrame
betas.columns = new_columns
```

```python
# Retain the first column and those columns matching df["Sample_ID"]
filtered_columns = [betas.columns[0]] + list(betas.columns[1:].intersection(df["Sample_ID"]))
filtered_betas = betas.loc[:, filtered_columns]

# Check for duplicate columns and remove them
filtered_betas = filtered_betas.loc[:, ~filtered_betas.columns.duplicated()]

# Verify the result
#print(f"Number of columns retained after filtering and removing duplicates: {filtered_betas.shape[1]}")

# Set the first column as the index
filtered_betas.set_index(filtered_betas.columns[0], inplace=True)

# Rename the index to "cpgs"
filtered_betas.index.name = "cpgs"

# Transpose the DataFrame
transposed_betas = filtered_betas.T

transposed_betas.index.name = "samples"

transposed_betas

```

```python
#align rows for spiec easi use

species = pd.read_csv("../ML_Data/FINAL_SPECIES_RAW.csv")
```

```python
# Check if the order of the rows in transposed_betas matches the order in species["Sample_ID"]
row_order_matches = list(transposed_betas.index) == list(species["Sample_ID"])

# Output the result
if row_order_matches:
    print("The row order in transposed_betas matches the order of Sample_ID in species.")
else:
    print("The row order in transposed_betas does NOT match the order of Sample_ID in species.")

# Optionally, identify mismatched rows if the order doesn't match
if not row_order_matches:
    for i, (trans_index, species_id) in enumerate(zip(transposed_betas.index, species["Sample_ID"])):
        if trans_index != species_id:
            print(f"Mismatch at position {i}: transposed_betas index = {trans_index}, species Sample_ID = {species_id}")
```

```python
# Align transposed_betas rows to match the order in species["Sample_ID"]
aligned_betas = transposed_betas.reindex(species["Sample_ID"])

# Verify the alignment
if list(aligned_betas.index) == list(species["Sample_ID"]):
    print("The rows of transposed_betas are now aligned with species['Sample_ID'].")
else:
    print("The alignment failed. Please check the data.")

# Output the first few rows of the aligned DataFrame
aligned_betas
```

```python
transposed_betas.to_csv("aligned_betas.csv", index=False)
```
